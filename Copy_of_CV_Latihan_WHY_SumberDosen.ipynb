# 3. Konversi Warna Pada Citra

```python
import cv2 as cv
from google.colab.patches import cv2_imshow #display sebuah gambar
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
import numpy as np

h=150
w=300
image=io.imread('/content/drive/MyDrive/Colab Notebooks/testing.jpg')
image=cv.cvtColor(image,cv.COLOR_BGR2RGB)
image=cv.resize(image,(w,h))
#cv2_imshow(image)

# Konversi RGB to Grayscale
gray = np.zeros((h, w), dtype=np.uint8)
for i in range(0,h):
  for j in range(0,w):
    gray[i][j]=int(image[i][j][0]*0.2126 + image[i][j][1]*0.7152 + image[i][j][2] * 0.0722)
#cv2_imshow(gray)

# Konversi RGB to CMY
cmy = np.zeros((h, w,3), dtype=np.uint8)
for i in range(0,h):
  for j in range(0,w):
    c = 255 - image[i][j][0]
    m = 255 - image[i][j][1]
    y = 255 - image[i][j][2]
    min_cmy = min(c, m, y)
    cmy[i][j]=[(c - min_cmy), (m - min_cmy), (y - min_cmy)]
#cv2_imshow(cmy)
```

# 4a. Constrast Stretching

```python
import cv2 as cv
from google.colab.patches import cv2_imshow #display sebuah gambar
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
import numpy as np

h, w=256, 256
image=io.imread('/content/drive/MyDrive/Colab Notebooks/lowcontrastgelap3.png')
image=cv.resize(image,(w,h))
#image=cv.cvtColor(image,cv.COLOR_BGR2GRAY)

sourceMin=image.min()
sourceMax=image.max()

a, b = 0, 255
output = np.zeros((h, w), dtype=np.uint8)
for i in range(0,h):
  for j in range(0,w):
    output[i][j]=(image[i][j]-sourceMin)*((b-a)/(sourceMax-sourceMin))+a

cv2_imshow(image)
cv2_imshow(output)
```

# 5. Histogram Equalization

```python
import cv2 as cv
from google.colab.patches import cv2_imshow #display sebuah gambar
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
import numpy as np
import scipy.stats as st

h, w=256, 256
image=cv.imread('/content/drive/MyDrive/Colab Notebooks/lowcontrastgelap3.png')
image=cv.resize(image,(w,h))

def draw_hist(x_axis, input):
    figsize = (5, 5)
    fig, ax = plt.subplots(figsize=figsize)
    plt.bar(x_axis, input, width=input.shape[0] / (x_axis[-1] - x_axis[0] + 1))
    return fig, ax

#Menghitung freqensi setiap level
bins_edges_min_max = [0, 255]
num_bins = 256
bin_count, bins_edges = np.histogram(image, num_bins, bins_edges_min_max)
bins_start = bins_edges[:-1]
#Menghitung probability
pdf = bin_count / np.sum(bin_count)
cdf = np.cumsum(pdf)
#Mengalikan dengan tujuan gambar
freq = np.round(cdf * 255).astype(int)
output = freq[image]

cv2_imshow(image)
#draw_hist(bins_start, bin_count)
#plt.title("Original histogram")
#plt.show()

cv2_imshow(output)
#bin_count, bins_edges = np.histogram(output, num_bins, bins_edges_min_max)
#draw_hist(bins_start, bin_count)
#plt.title("Histogram Hasil HE")
#plt.show()
```


# 6. Enhancement with Filtering

```python
import cv2 as cv
from google.colab.patches import cv2_imshow #display sebuah gambar
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
import numpy as np
import scipy.stats as st

h, w=360, 480
image=io.imread('/content/drive/MyDrive/Colab Notebooks/Blur.jpg')
image=cv.cvtColor(image,cv.COLOR_BGR2GRAY)
#image=cv.resize(image,(w,h))

def konvolusi(input_img, kernel):
  #inisialisasi output dengan matrix dengan nilai 0
  output = np.zeros(input_img.shape)
  #padding angka 0 pada border input gambar
  padding = np.pad(input_img, pad_width=1, mode='constant', constant_values=0)
  for i in range(input_img.shape[0]):
    for j in range(input_img.shape[1]):
      for k in range(kernel.shape[0]):
        for l in range(kernel.shape[1]):
          output[i][j] = output[i][j] + kernel[k][l]*padding[i+k][j+l]
  return output

output = np.zeros(image.shape);
kernel = np.array([[-0.11, -0.11, -0.11],
                    [-0.11, 1.89, -0.11],
                    [-0.11, -0.11, -0.11]])
output = konvolusi(image, kernel)
#output2 = cv.filter2D(image,-1,kernel)

cv2_imshow(image)
cv2_imshow(output)
```

# 9. Background Substraction

```python
import cv2 as cv
from google.colab.patches import cv2_imshow #display sebuah gambar
from PIL import Image
import numpy as np

h, w=256, 256
background=cv.imread('/content/drive/MyDrive/Colab Notebooks/Jalan0.jpg')
background=cv.cvtColor(background,cv.COLOR_BGR2GRAY)
background=cv.resize(background,(w,h))

image=cv.imread('/content/drive/MyDrive/Colab Notebooks/Jalan1.jpg')
image=cv.cvtColor(image,cv.COLOR_BGR2GRAY)
image=cv.resize(image,(w,h))

diff = np.zeros((h, w), dtype=float)
binimage = np.zeros((h, w), dtype=float)
for i in range(0,h):
  for j in range(0,w):
    diff[i][j]=abs(float(image[i][j])-float(background[i][j]))
    if (diff[i][j]>40):
      binimage[i][j]=255;
    else:
      binimage[i][j]=0;

cv2_imshow(background)
cv2_imshow(image)
cv2_imshow(diff)
cv2_imshow(binimage)
```

# 10. Object Classification Using CNN

```python
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.layers import Dense
from keras.preprocessing.image import ImageDataGenerator

classifier = Sequential()
classifier.add(Conv2D(32, (3,3),input_shape=(64,64,3),activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2,2)))
classifier.add(Conv2D(32, (3,3),activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2,2)))

classifier.add(Flatten())
classifier.add(Dense(units=128, activation='relu'))
classifier.add(Dense(units=1, activation='sigmoid'))

classifier.compile(optimizer='adam',loss = 'binary_crossentropy',metrics = ['accuracy'])

train_datagen=ImageDataGenerator(rescale=1./255,shear_range = 0.2,zoom_range = 0.2,horizontal_flip = True)
test_datagen = ImageDataGenerator(rescale=1./255)
training_set = train_datagen.flow_from_directory('/content/drive/MyDrive/Colab Notebooks/cnn/training_set',
                                                 target_size = (64,64),
                                                 batch_size=32,
                                                 class_mode='binary')
test_set = test_datagen.flow_from_directory('/content/drive/MyDrive/Colab Notebooks/cnn/testing_set',
                                            target_size=(64,64),
                                            batch_size=32,
                                            class_mode = 'binary')

classifier.fit(training_set, steps_per_epoch = 2, epochs = 10, validation_data = test_set, validation_steps=10)
```

# 11. Face Recognition

```python
from google.colab.patches import cv2_imshow
import cv2 as cv
import os
import numpy as np
from keras.preprocessing import image

def detect_face(img):
  gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
  face_cascade = cv.CascadeClassifier("/content/drive/MyDrive/Colab Notebooks/haarcascade_frontalface_default.xml")
  faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=5)
  if (len(faces)==0):
    return None, None
  (x,y,w,h)=faces[0]
  return gray[y:y+w, x:x+h], faces[0]

def prepare_training_data(data_folder_path):
  dirs = os.listdir(data_folder_path)
  faces = []
  labels = []
  
  for dir_name in dirs:
    if not dir_name.startswith("s"):
      continue;
    label = int(dir_name.replace("s", ""))
    subject_dir_path = data_folder_path + "/" + dir_name
    subject_images_names = os.listdir(subject_dir_path)
    for image_name in subject_images_names:
      if image_name.startswith("."):
        continue;
      image_path = subject_dir_path + "/" + image_name
      image = cv.imread(image_path)
      face, rect = detect_face(image)
      if face is not None:
        face=cv.resize(face, (50, 50))
        faces.append(face)
        labels.append(label)
  return faces, labels

def draw_rectangle(img, rect):
  (x, y, w, h) = rect
  cv.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)

def draw_text(img, text, x, y):
  cv.putText(img, text, (x, y), cv.FONT_HERSHEY_PLAIN, 1.5, (0, 255, 0), 2)

faces, labels = prepare_training_data('/content/drive/MyDrive/Colab Notebooks/training-data')
```

```python
subjects = ["", "Jokowi", "Prabowo"]
face_recognizer = cv.face.FisherFaceRecognizer_create()
face_recognizer.train(faces,np.array(labels))
```

```python
test_image=cv.imread('/content/drive/MyDrive/Colab Notebooks/testing-data/prabowo.jpg');
img=test_image.copy();
face, rect = detect_face(img);
face=cv.resize(face, (50, 50));
label = face_recognizer.predict(face);
label_text = subjects[label[0]];
draw_rectangle(img,rect);
draw_text(img, label_text, rect[0], rect[1]-5);

cv2_imshow(img);
```
