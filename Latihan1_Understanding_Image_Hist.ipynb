```python
image = cv.imread('/content/drive/MyDrive/Colab Notebooks/img/asal.jpg')
b ,g, r = cv.split(image)#membagikan nilai 3 array menjadi 3 nilai variable
cv2_imshow( image)
# cv2_imshow("biru",b) ga bisa diberi nama var
cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)
```

```python

plt.hist(b.ravel(),256,[0,256])
plt.hist(g.ravel(),256,[0,256])
plt.hist(r.ravel(),256,[0,256])
plt.show()#jangan lupa tampilkan
```

```python
rgb = (b,g,r)
warna =['b','g','r']
for i,col in enumerate(rgb):
  plt.hist(col.ravel(),256,[0,256],color=warna[i])
  plt.show()
```

```python
image = cv.imread('/content/drive/MyDrive/Colab Notebooks/img/asal.jpg',0)
#menggunakan cv.histCalc method
hist = cv.calcHist([image],[0],None,[256],[0,256] )#image,channel, mask,histSize,range, hist=None
plt.plot(hist)
```

```python
#membuat citra sndiri (rectangle)
img = np.zeros((256,256),np.uint8)
cv.rectangle(img, (0,128),(256,256),(100),-1) #img,pt1<horizontal>, pt2, di nilai x berapa, linetype
cv.rectangle(img, (2,64),(128,128),(150),-1) #img,pt1(horizontal(posisi awal,tinggi)), pt2(vertikal(lebar,akhir posisi)), di nilai x berapa(warna), linetype
plt.hist(img.ravel(),256,[0,256])# x, jumlah bin, range(nilai sumbu x)
plt.show()#jangan lupa tampilkan
# DisabledFunctionError: cv2.imshow() is disabled in Colab, because it causes Jupyter sessions
# to crash; see https://github.com/jupyter/notebook/issues/3935.
# As a substitution, consider using
#   from google.colab.patches import cv2_imshow
#di google collab ada library khusus untuk menampilkan gambar open cv,, jika mengggunakan
#hanya cv.imshow, tidak bisa, karna tidak ada di gcollab
cv2_imshow(img)
```
