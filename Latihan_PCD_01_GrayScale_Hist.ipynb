# Image Histogram

```python
from PIL import Image
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
%matplotlib inline
```

Pertama kita baca gambar terlebih dahulu

```python

image = cv.imread('/content/drive/MyDrive/Colab Notebooks/img/asal.jpg')
cv2_imshow(image)
```

```python
image = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
cv2_imshow(image)
```

```python
print(image)
```

```python
(n,m)=(image.shape)
# print(n)
# print(m)
k = 0
H = np.zeros((256), dtype = int)
# print(H)

while k <256:
  H[k] =np.count_nonzero(image == k)  #(image)
  k = k+1
  
intensity = np.arange(0,256,1)

#x,y,color,width
plt.bar(intensity, H, color ='blue', width = 0.5) 
plt.xlabel("Intensity")
plt.ylabel("Frequency")
plt.title("Histogram")


plt.show()

print(H)
```

```python
# NO 1 (HISTOGRAM DARI IMAGE)
grayimg = image.astype(int)
keabuan = np.zeros(256,dtype=int)
r = np.zeros(256,dtype=int)
for x in range (256):
    r[x] = x
tinggi, lebar = grayimg.shape
n = tinggi*lebar
for i in range(tinggi):
    for j in range (lebar):
        keabuan[ grayimg[i,j] ] += 1
plt.xlabel('Keabuan')
plt.ylabel('Jumlah')
plt.bar(r, keabuan,color ='blue', width = 0.5)
plt.show()
```

```python
plt.clf
plt.title("histogram")
hist = np.histogram(image, bins=np.arange(0,256))
plt.plot(hist[1][:-1], hist[0], lw=2)
```

#IMAGE TRANSFORM GRAYSCALE

```python
#bright pixel became dark, dark became bright
gray_image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
#cv2_imshow(gray_image)
invert_gray = 255 - gray_image
cv2_imshow(invert_gray)
```

```python
plt.hist(invert_gray.ravel(), bins = 256, range = [0,256])
plt.show()
```

#High Brightness Image

adding more bright, hazing-like image

analisis histogram dari gambar grayscale yang memiliki intensitas terang berlebih

```python
bright_image = (100.0/255)*gray_image + 100
cv2_imshow(bright_image)
```

```python
plt.hist(bright_image.ravel(), bins = 256, range = [0,256])
plt.show()
```

#DARKEN IMAGE TRANSFORM

The Lightness level decreased

analisis histogram dari gambar grayscale yang telah diubah tingkat gelapnya ditambahkan

```python
darken_image = 255.0*(gray_image/255.0)**2
cv2_imshow(darken_image)
```

```python
plt.hist(darken_image.ravel(), bins = 256, range = [0,256])
plt.show()
```
